```python
from gtts import gTTS
import IPython.display as ipd
```

```python
# Function to convert text to speech and play audio
def text_to_speech(text):
    # Create gTTS object with Brazilian Portuguese language
    tts = gTTS(text=text, lang='pt', tld='com.br')

    # Save the audio to a temporary file
    audio_file = "output.mp3"
    tts.save(audio_file)

    # Play the audio in Colab
    return ipd.Audio(audio_file)

# Example usage
text = "No princípio era a Palavra e a Palavra estava com Deus e a Palavra era Deus. Aquele que é a Palavra sempre esteve com Deus. Criou tudo o que existe e nada existe que não tenha sido feito por ele. "
audio = text_to_speech(text)
audio
```

```python
from googletrans import Translator

def translate_text(text, dest_language='en'):
    translator = Translator()
    translated_text = translator.translate(text, dest=dest_language)
    return translated_text.text

# Example usage with the original Portuguese text
text_portuguese = "No princípio era a Palavra e a Palavra estava com Deus e a Palavra era Deus. Aquele que é a Palavra sempre esteve com Deus. Criou tudo o que existe e nada existe que não tenha sido feito por ele."
text_english = translate_text(text_portuguese, dest_language='en')
print(text_english)
```
