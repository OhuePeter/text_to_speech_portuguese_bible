```python
from gtts import gTTS
import IPython.display as ipd
```

```python
# Function to convert text to speech and play audio
def text_to_speech(text):
    # Create gTTS object with Brazilian Portuguese language
    tts = gTTS(text=text, lang='pt', tld='com.br')

    # Save the audio to a temporary file
    audio_file = "output.mp3"
    tts.save(audio_file)

    # Play the audio in Colab
    return ipd.Audio(audio_file)

# Example usage
text = "São estes os antepassados de Jesus Cristo, descendente do rei David e de Abraão."
audio = text_to_speech(text)
audio
```
